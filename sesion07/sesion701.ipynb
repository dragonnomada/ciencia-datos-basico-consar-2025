```python
seaborn.kdeplot(x=titanic["Fare"], hue=titanic["Sex"])
```

```python
seaborn.boxplot(y=titanic["Fare"], hue=titanic["Sex"], showfliers=False)
```

```python
titanic.iloc[4]["Fare"]
```

```python
import matplotlib.pyplot as pyplot

seaborn.boxplot(y=titanic["Fare"], 
    hue=titanic["Survived"].map({
        0: "No sobrevive",
        1: "Sobrevive"
    }), 
    showfliers=False)

fare = titanic.iloc[4]["Fare"]

pyplot.plot([-0.4, 0], [fare, fare], color="red")
pyplot.title("Comparación de tarifas y supervivencia \n en los pasajeros del Titanic")
pyplot.xlabel("Supervivencia")
pyplot.ylabel("Tarifa")
pyplot.legend()
```
