Al analizar las plantas iris debemos centrarnos en entender el comportamiento poblacional a través de una muestra de datos. De las muestras tendremos características numéricas y categóricas que exhibirán el comportamiento estadístico/matemático del problema.

### Análisis de contraste

Para poder construir un análisis de contraste necesitamos una serie de datos numérica que sea sometida a la hipótesis de que tiene un comportamiento distinto en diferentes grupos, a esta serie le llamaremos eje de análisis.

Para formar los grupos necesitamos una segunda serie de datos categórica que representará un grupo por cada categoría, a esta serie le llamaremos eje de contraste.

```python
import seaborn

seaborn.kdeplot(x=iris["sepal.length"], hue=iris["variety"])
```

![iris-sepalo-largo](iris-sepalo-largo.png)

```python
seaborn.kdeplot(x=iris["petal.length"], hue=iris["variety"])
```

```python
seaborn.boxplot(y=iris["sepal.length"], hue=iris["variety"])
```

```python
seaborn.boxplot(y=iris["petal.length"], hue=iris["variety"])
```

### Conclusiones

Al analizar la planta iris se detectan tres familias o variedades cuyo comportamiento es distinto en los largos del pétalo y sépalo. Esto reafirma la hipótesis de que la variedad influye en las medidas del pétalo y sépalo, por lo que se puede aprender el comportamiento de los datos a través de su variedad.

Esto significa que somos capaces de poder determinar una variedad de la planta iris con alta presición para setosa y un poco menos precisión para determinar si es virgínica o versicolor. Es decir, podemos clasificar la variedad de las plantas iris a partir de sus características.
