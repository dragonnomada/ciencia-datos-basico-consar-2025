## Problemática

Para analizar el conjunto de datos sobre la supervivencia de los pasejeros del Titanic, tenemos:

**Población**

Los pasejeros que viajaron en el Titanic.

**Muestra**

891 registros con las características sobre los pasejeros del Titanic.

**Características**

* Supervivencia - Categórico (2 clases / binario) 
    * `1` - Sobrevive
    * `0` - No sobrevive
* Clase - Categórico (3 clases)
    * `1` - Primera clase
    * `2` - Segunda clase
    * `3` - Tercera clase
* Sexo - Categórico (3 clases)
    * `male` - Hombre
    * `female` - Mujer
* Edad - Numérica
* Familiares - Numérico
* Padre o hijo - Numérico
* Tarifa - Numérico
* Puerto - Categórico (3 clases)

**Hipótesis**

1. La tarifa (numérico) tiene un comportamiento distinto por supervivencia (categórico)
2. La edad (numérico) tiene un comportamiento distinto por supervivencia (categórico)
3. La tarifa (numérico) tiene un comportamiento distinto por puerto (categórico)
4. La tarifa (numérico) tiene un comportamiento distinto por sexo (categórico)

**Estadítica**

Comportamiento de la media y dispersión por grupos mediante un análisis de contraste.
