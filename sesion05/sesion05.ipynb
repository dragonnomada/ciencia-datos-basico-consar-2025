## Tablas de datos (DataFrame)

Una serie de datos son observaciones de una misma característica, pero cuando analizamos varias características se formarán distintas series de datos, para no manipular serie por serie, la tabla de datos concentra múltiples series en un mismo objeto como una hoja de Excel, en sentido de poder concentrar los registros de diferentes características sobre un mismo elemento.

Entonces, una tabla de datos, puede ser vista como una entidad que descompone en filas y columnas, donde cada columna representa una serie de datos y por ende una característica de la entidad, mientras que cada fila reprensentará una observación sobre todas las características.

Si analizamos por fila, obtendremos la información de cada observación y si analizamos por columna, obtendremos la información de cada característica.

Por ejemplo, suponga la siguiente tabla de datos sobre dos características de una muestra de personas.

### Conteo apilado (stack count)

Cuando tenemos múltiples series categóricas, podemos formar un conteo conocido como el conteo apilado por formato lago.

El forma largo se refiere a que los datos están desglosados como filas combinadas y el formato ancho transforma o agrupa estos conteos en forma de matriz.

Para lograrlo, primero se debe dominar el formato largo y luego traducirlo mediante una agrupación al formato ancho.

Para visualizar los conteos se puede utilizar un mapa de calor.

> Conteos en formato largo

```python
personas.groupby(["Género", "Color Cabello"]).size()
```

```python
personas.groupby(["Color Cabello", "Género"]).size()
```

> Conteos en formato ancho

```python
personas.groupby(["Género", "Color Cabello"]).size().unstack()
```

```python
personas.groupby(["Color Cabello", "Género"]).size().unstack()
```

> Ejemplo, probabilidades marginales y combinadas

```python
personas["Género"].value_counts() / len(personas)
```

```python
personas["Color Cabello"].value_counts() / len(personas)
```

```python
personas.groupby(["Género", "Color Cabello"]).size().unstack() / len(personas)
```

```python
reporte1 = personas.groupby(["Género", "Color Cabello"]).size().unstack() / len(personas)

reporte1
```

Sumar sobre el eje 0, significa obtener cada columna con la suma de todas sus filas.

```python
reporte1.sum(axis=0) # Devuelve la suma las filas / Columnas reducidas 
```

Sumar sobre el eje 1, significa obtener cada fila con la suma de todas sus columnas.

```python
reporte1.sum(axis=1) # Devuelve las suma de las columnas / Filas reducidas
```

Reporte con marginales

```python
reporte2 = reporte1.copy()

reporte2["Total"] = reporte2.sum(axis=1)
reporte2.loc["Total"] = reporte2.sum(axis=0)

reporte2
```

### Caso de estudio - Probabilidades condicionales 

$$
P(Hombre | Castaño) = \frac{P(Hombre, Castaño)}{P(Castaño)}
$$

```python
reporte1.loc["Hombre", "Castaño"]
```

```python
reporte1.sum(axis=0).loc["Castaño"]
```

```python
reporte1.loc["Hombre", "Castaño"] / reporte1.sum(axis=0).loc["Castaño"]
```

$$
P(Rubio | Hombre) = \frac{P(Hombre, Rubio)}{P(Hombre)}
$$

```python
reporte1.loc["Hombre", "Rubio"] / reporte1.sum(axis=1).loc["Hombre"]
```
