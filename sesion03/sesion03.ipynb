## Graficación con Matplotlib

Instalación de la librería

Importación de la librería

```python
import matplotlib.pyplot as pyplot
```

Defininición de una serie numérica (edades de personas en algún grupo o población)

```python
edades = [23, 17, 85, 64, 35, 21, 14, 19]
```

Gráfica de histograma

```python
pyplot.hist(edades)
pyplot.show()
```

Gráfica de histograma con 4 segementos (*bins*)

```python
pyplot.hist(edades, bins=4)
pyplot.show()
```

Gráfica de caja (caja estadística)

```python
pyplot.boxplot(edades, vert=False)
pyplot.show()
```

Gráfica de caja sin puntos atípicos (outliers / fliers) 

```python
pyplot.boxplot(edades, vert=False, showfliers=False)
pyplot.show()
```

Gráfica de violín (o mariposa)

```python
pyplot.violinplot(edades, vert=False)
pyplot.show()
```
